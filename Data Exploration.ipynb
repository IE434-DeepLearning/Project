```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
```

```python
debug = False
if debug:
    file_name = 'debug_data.pkl'
else:
    file_name = 'data.pkl'
```

```python
df = pd.read_pickle('./Data/Rawdata/' + file_name)
df.head()
```

```python
# These are the feature to keep from the project description provided
columns_to_keep = ['borough', 'job_type', 'block', 'lot', 'zip_code', 'work_type', 'bldg_type', "owner's_business_type",
                   'non-profit', 'latitude', 'longitude', 'council_district', 'issuance_date', 'job_start_date',
                   'permit_type']
df = df[columns_to_keep]
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.reset_index(inplace=True, drop=True)
df.info()
```

```python
df['block'] = df['block'].astype('category')
df['lot'] = df['lot'].astype('category')
df['zip_code'] = df['zip_code'].astype('category')
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df['permit_type'].replace('AL', 'Alteration', inplace=True)
df['permit_type'].replace('DM', 'Demolition', inplace=True)
df['permit_type'].replace('EQ', 'Construction Equipment', inplace=True)
df['permit_type'].replace('EW', 'Equipment Work', inplace=True)
df['permit_type'].replace('FO', 'Foundation', inplace=True)
df['permit_type'].replace('NB', 'New Building', inplace=True)
df['permit_type'].replace('PL', 'Plumbing', inplace=True)
df['permit_type'].replace('SG', 'Sign', inplace=True)
```

```python
df.permit_type.value_counts()
```

```python
#plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='permit_type')
plt.title("Distribution of Permit Types", fontsize = 20)
plt.xlabel('Permit Type', fontsize = 15)
plt.ylabel("Count", fontsize = 15)
plt.xticks(rotation=45)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

#plt.savefig('Distribution of Permit Types.png')
plt.show()
```

```python
categorical_features = ['borough', 'job_type', 'work_type', 'bldg_type', "owner's_business_type", 'non-profit',
                        'council_district']

for feature in categorical_features:
    print(df[feature].value_counts(), end='\n\n')
    df[feature].value_counts().plot(kind='bar')
    plt.show()
    print('\n\n\n')
```

```python
df.zip_code.value_counts()
```

```python
df['issuance_year'] = df.issuance_date.dt.year
df['issuance_month'] = df.issuance_date.dt.month
df['issuance_day'] = df.issuance_date.dt.day

df['job_start_year'] = df.job_start_date.dt.year
df['job_start_month'] = df.job_start_date.dt.month
df['job_start_day'] = df.job_start_date.dt.day
```

```python
correlation_matrix = df.apply(lambda x: pd.factorize(x)[0]).corr(method='pearson', min_periods=1)
correlation_matrix
```

```python
correlation_matrix.columns = ["Borough", "Job Type", "Block", "Lot", "Zip Code", "Work Type", "Bulding Type", "Owner's Business Type", "Non-Profits",
                            "Latitude", "Longitude", "Council District", "Issuance Date", "Job Start Date", "Permit Type", "Issuance Year",
                            "Issuance Month", "Issuance Day", "Job Start Year", "Job Start Month", "Job Start Day"]
correlation_matrix.index = ["Borough", "Job Type", "Block", "Lot", "Zip Code", "Work Type", "Bulding Type", "Owner's Business Type", "Non-Profits",
                            "Latitude", "Longitude", "Council District", "Issuance Date", "Job Start Date", "Permit Type", "Issuance Year",
                            "Issuance Month", "Issuance Day", "Job Start Year", "Job Start Month", "Job Start Day"]
 
plt.figure(figsize=(16, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5, annot_kws={"size": 12})
plt.title('Correlation Heatmap', fontsize=16)
plt.xticks(fontsize=12)  # for x-axis
plt.yticks(fontsize=12)  # for y-axis
plt.show()
#plt.savefig("Correlation Matrix.png", dpi=300)
```

```python
# Deal with string values in 'block', 'lot', and 'zip_code'
# Assuming df is your DataFrame
df['block'] = pd.to_numeric(df['block'], errors='coerce')
df['lot'] = pd.to_numeric(df['lot'], errors='coerce')
df['zip_code'] = pd.to_numeric(df['zip_code'], errors='coerce')
 
# Detect rows with non-numeric values
non_numeric_blocks = df[df['block'].isna()]
non_numeric_lots = df[df['lot'].isna()]
non_numeric_zip_codes = df[df['zip_code'].isna()]
 
# Print or inspect the rows with non-numeric values
print("Rows with non-numeric values in 'block':")
print(non_numeric_blocks)
print("\nCount of rows with non-numeric values in 'block':", len(non_numeric_blocks))
 
print("\nRows with non-numeric values in 'lot':")
print(non_numeric_lots)
print("\nCount of rows with non-numeric values in 'lot':", len(non_numeric_lots))
```
