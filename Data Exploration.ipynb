# Before addressing the null values, we will drop the columns which will not add any value to the model training process and which might have a lot of unique values.

### We decided to drop the string columns using the above table containing the number of unique values in each column. The string columns were also assigned based on this criterion i.e. number of unique values in the column.

```python
columns_to_drop = ['bin_#', 'house_#', 'street_name', 'job_#', ]
```

```python
col_dtypes = list(zip(list(df.columns), list(map(lambda x: str(type(df[x].dtype)), list(df.columns)))))
```

```python
columns_to_drop = [x[0] for x in col_dtypes if 'StringDtype' in x[1]]
columns_to_drop
```

```python
df_col_dropped = df.drop(columns_to_drop, axis=1)
df_col_dropped
```
